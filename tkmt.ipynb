```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
pl = pd.read_csv('placeQN.csv')
pl.head()
```

```python
pl.describe()
```

```python
na = len(pl[pl["Place's total reviewer"]==0])
print('Tỉ lệ dữ liệu trống', na/len(pl)*100,"%")
```

```python
df_point = pl[["Place's 5 points","Place's 4 points","Place's 3 points","Place's 2 points","Place's 1 points"]]
df_point
```

```python
list = []
for i in df_point:
    list.append(sum(df_point[i]))
print(list)
name = ['5 Sao','4 Sao','3 Sao','2 Sao','1 Sao']
fig = plt.figure(figsize =(10, 7))
plt.pie(list, labels = name)
plt.title('Tỉ lệ đánh giá địa điểm du lịch tại Quảng Nam')
plt.legend()
plt.show()
```

```python
dict_columns = {
    'Đánh giá':['5 Sao', '4 Sao', '3 Sao', '2 Sao', '1 Sao'],
    'Tổng số đánh giá':[124057, 23185, 9212, 2785, 2655]
}
df_sao = pd.DataFrame(dict_columns)
df_sao
```

```python
fig = plt.figure(figsize =(15, 8))
ax = sns.barplot(x="Đánh giá", y="Tổng số đánh giá", data=df_sao)
```

```python
a = pl.groupby("Place's type").size().reset_index(name='Số địa điểm')
h = pd.DataFrame(a)
print("Số thể loại địa điểm giải trí tại Quảng Nam là: ", len(h))
print("Top 10 thể loại giải trí tại Quảng Nam")
df_type = h.nlargest(10, 'Số địa điểm')
print(df_type)
fig = plt.figure(figsize =(10, 5))
ax = sns.barplot(x="Số địa điểm", y="Place's type", data=df_type)
```

```python
print("Top 10 địa điểm có nhiều du khách tham quan nhất tại Quảng Nam là:")
df10 = pl.nlargest(10, "Place's total reviewer")[["Place's name","Place's total reviewer"]]
print(df10)
fig = plt.figure(figsize =(10, 5))
ax = sns.barplot(x="Place's total reviewer", y="Place's name", data=df10)
```

```python
print('Top 10 địa điểm có nhiều ảnh nhất tại Quảng Nam')
df_pic = pl.nlargest(10, "Place's images")[["Place's name","Place's images"]]
df_pic
fig = plt.figure(figsize =(10, 5))
ax = sns.barplot(x="Place's images", y="Place's name", data=df_pic)
```

```python
rv = pd.read_csv('reviewQN.csv')
rv.head()
```

```python
rv.describe()
```

```python
rv_name = rv.groupby("Reviewer's name").size().reset_index(name='Số comment')
rv_name
rv_name = pd.DataFrame(rv_name)
print("Số khách đã comment tại Quảng Nam là: ", len(rv_name))
print("Top 10 khách hàng bình luận nhiều nhất")
df_rv = rv_name.nlargest(10, 'Số comment')
df_rv
```

```python
rv_c = rv.groupby("Reviewer's name")["Reviewer's contribution"].sum().reset_index(name="Số đóng góp")
rv_c = pd.DataFrame(rv_c)
print("Top 10 khách hàng đóng góp nhiều nhất")
df_rv = rv_c.nlargest(10, 'Số đóng góp')
df_rv
```

```python
rv["Reviewers's time feedback"] = pd.to_datetime(rv["Reviewers's time feedback"])
```

```python
rv_date = rv.groupby("Reviewers's time feedback")["url"].size().reset_index(name="Số comment")
rv_date = pd.DataFrame(rv_date)
plt.figure(figsize=(10,8))
plt.plot(rv_date["Reviewers's time feedback"],rv_date["Số comment"])
plt.title('BIỂU ĐỒ THỂ HIỆN SỐ COMMENT TẠI QUẢNG NAM QUA CÁC NĂM')
plt.ylabel("Số comment")
plt.show()
```

Nhìn vào biểu đồ thì ta thấy rằng số comment tăng mạnh từ năm 2016 đến năm 2020.
Từ 2020 trở đi thì số comment đã giảm mạnh đáng kể do dịch Covid-19.
